PROBLEM OBJECTIVE: ONLINE DELIVERY

(1)DETERMINE THE PERCENTAGE OF RESTAURANTS THAT OFFER ONLINE DELEVERY.

(2)COMPARE THE AVERAGE RATINGS OF RESTAUTANTS WITH AND WITHOUT ONLINE DELEVERY.


```python
#(1)DETERMINE THE PERCENTAGE OF RESTAURANTS THAT OFFER ONLINE DELEVERY..

percentahe_online_delivery = data['Has Online delivery'].value_counts(normalize = 'true')*100
percentahe_online_delivery
```

```python
plt.figure(figsize=(6,4))
colors = ['red', 'green']
plt.bar(percentahe_online_delivery.index, percentahe_online_delivery , color=colors)
plt.title('Online Delivery')
plt.show()

```

```python
#(2)COMPARE THE AVERAGE RATINGS OF RESTAUTANTS WITH AND WITHOUT ONLINE DELEVERY.

compare_average_rat = data.groupby('Has Online delivery')['Aggregate rating'].mean()
compare_average_rat

```
